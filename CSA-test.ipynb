```python
from numpy import around, random, sqrt, exp, append, linspace, log, arange, size, zeros, identity, ones, log, sin, cos, tanh, pi, clip
from pylab import figure, clf, show, xlabel, ylabel, rcParams, pcolor, imshow, cm, clim, meshgrid, colorbar
from matplotlib.pyplot import close
from bisect import bisect_right, bisect_left

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import bqplot
import bqplot.marks as bqm
import bqplot.scales as bqs
import bqplot.axes as bqa
from ipywidgets import FloatSlider,interactive,widgets,interact,Label
from IPython.display import display
import bqplot.pyplot as plt

from timeit import default_timer as timer
start = timer()

def gen_E1(X,Z,d,k1):
    global E1
    z0=2.5
    X1=X-d/2
    E1=exp(2.j*pi*sqrt(X1*X1+Z*Z)*k1)/(k1*sqrt(X1*X1+Z*Z))
    return E1

#def gen_E2(X,Z,d,k2):
def gen_E2(X,Z,d,k1):
    global E2
    X1=X+d/2
#    E2=exp(2.j*pi*sqrt(X1*X1+Z*Z)*k2)/(k2*sqrt(X1*X1+Z*Z))
    E2=exp(2.j*pi*sqrt(X1*X1+Z*Z)*k1)/(k1*sqrt(X1*X1+Z*Z))
    return E2

def make_map_1(E1,E2):
    E_tot=E1+E2
    Int=abs(E_tot)**2
    Int=log(Int)
    map1=bqm.HeatMap(x=z,y=x,color=Int,scales=sc_plot)
    return map1

def E1_updtr(change):
    global d
    d=sep_sldr.value
    global k1
    k1=k1_sldr.value
    global E1
    E1=gen_E1(X,Z,d,k1)
    w1=gen_w1(nlin,k1,d)
    if ln_hide==0:
        fig_1.marks=([make_map_1(E1,E2)])
    elif ln_hide==1:
        fig_1.marks=([make_map_1(E1,E2)]+list(w1_lin)+list(w2_lin))
    
def E2_updtr(change):
    global d
    d=sep_sldr.value
#    global k2
    global k1
#    k2=k2_sldr.value
    k1=k1_sldr.value
    global E2
#    E2=gen_E2(X,Z,d,k2)
    E2=gen_E2(X,Z,d,k1)
#    w2=gen_w2(nlin,k2,d)
    w2=gen_w2(nlin,k1,d)
    if ln_hide==0:
        fig_1.marks=([make_map_1(E1,E2)])
    elif ln_hide==1:
        fig_1.marks=([make_map_1(E1,E2)]+list(w1_lin)+list(w2_lin))
        

def E1_E2_updtr(change):
    global d
    d=sep_sldr.value
    global k1
    k1=k1_sldr.value
    global E1
    E1=gen_E1(X,Z,d,k1)
    w1=gen_w1(nlin,k1,d)
    global E2
    E2=gen_E2(X,Z,d,k1)
    w2=gen_w2(nlin,k1,d)
    if ln_hide==0:
        fig_1.marks=([make_map_1(E1,E2)])
    elif ln_hide==1:
        fig_1.marks=([make_map_1(E1,E2)]+list(w1_lin)+list(w2_lin))


def sep_updtr(change):
    global d
    d=sep_sldr.value
    global E1
    E1=gen_E1(X,Z,d,k1)
    global E2
#    E2=gen_E2(X,Z,d,k2)
    E2=gen_E2(X,Z,d,k1)
    w1=gen_w1(nlin,k1,d)
#    w2=gen_w2(nlin,k2,d)
    w2=gen_w2(nlin,k1,d)
    if ln_hide==0:
        fig_1.marks=([make_map_1(E1,E2)])
    elif ln_hide==1:
        fig_1.marks=([make_map_1(E1,E2)]+list(w1_lin)+list(w2_lin))
    
def hide_updtr(change):
    global ln_hide    
    if change["new"]==False:
        ln_hide=0
        fig_1.marks=([make_map_1(E1,E2)])
    elif change["new"]==True:
        ln_hide=1
        fig_1.marks=([make_map_1(E1,E2)]+list(w1_lin)+list(w2_lin))
        
def res_updtr(change):
    global dx
    global X
    global Z    
    if change["new"]==False:
        dx=200
    elif change["new"]==True:
        dx=800
    global x
    global z
    x=linspace(0,2*xmax,dx)
    z=linspace(-xmax,xmax,dx)
    Z,X = meshgrid(x, z)
    gen_E1(X,Z,d,k1)
#    gen_E2(X,Z,d,k2)
    gen_E2(X,Z,d,k1)
    if ln_hide==0:
        fig_1.marks=([make_map_1(E1,E2)])
    elif ln_hide==1:
        fig_1.marks=([make_map_1(E1,E2)]+list(w1_lin)+list(w2_lin))

xmax=5 # no. of x points, if dx is fixed, larger value gives, larger range
dx=200# 0.025 takes 120s

d=6.0
theta3=0
k1=1.0
#k2=k1
ln_hide=1

x=linspace(0,2*xmax,dx)
z=linspace(-xmax,xmax,dx)

Z,X = meshgrid(x, z)
gen_E1(X,Z,d,k1)
#gen_E2(X,Z,d,k2)
gen_E2(X,Z,d,k1)

sc_x = bqs.LinearScale(min=-xmax,max=xmax)
sc_y = bqs.LinearScale(min=0,max=2*xmax)
sc_col = bqplot.ColorScale(colors=['#000080','#FFFF00','#ff0000'],max=10)
sc_plot={'row': sc_x, 'column': sc_y, 'color': sc_col}

ax_x_1 = bqa.Axis(scale=sc_x , num_ticks=0)
ax_y_1 = bqa.Axis(scale=sc_y,  orientation='vertical',num_ticks=0)
ax_x_y_1=[ax_x_1,ax_y_1]


fig_layout_0=widgets.Layout(width='400px',height='400px')
margin_0=dict(top=-10, bottom=-10, left=0, right=0)

def gen_w1(nlin,k1,d):
    global w1_lin
    test1=5000
    w1_lin=[]
    nlin=int(2*sqrt(2)*xmax*k1+1)
    for ii in range(0,nlin):
        step=ii/k1       
        xt1=linspace(xmax+d/2-step,xmax+d/2+step,test1)
        xt=xt1[bisect_left(xt1,0):bisect_right(xt1, 2*xmax)]
        w1_lin=append(w1_lin,bqplot.marks.Lines(y=xt[1:int(test1-1)],
                                                x=(-xmax+sqrt((step)**2-(xt[1:int(test1-1)]-(xmax+d/2))**2)), 
                                                colors=['#FFFFFF'],
                                               scales={'x': sc_x, 'y': sc_y}))
    return list(w1_lin)

#def gen_w2(nlin,k2,d):
def gen_w2(nlin,k1,d):
    global w2_lin
    test1=5000
    w2_lin=[]
#    nlin=int(2*sqrt(2)*xmax*k2+1)
    nlin=int(2*sqrt(2)*xmax*k1+1)
    for ii in range(0,nlin):
#        step=ii/k2
        step=ii/k1
        xt1=linspace(xmax-d/2-step,xmax-d/2+step,test1)
        xt=xt1[bisect_left(xt1,0):bisect_right(xt1, 2*xmax)]
        w2_lin=append(w2_lin,bqplot.marks.Lines(y=xt[1:int(test1-1)],
                                                x=(-xmax+sqrt((step)**2-(xt[1:int(test1-1)]-(xmax-d/2))**2)), 
                                                colors=['#FFFFFF'],
                                                scales={'x': sc_x, 'y': sc_y}))
    return list(w2_lin)

nlin=int(2*sqrt(2)*xmax*k1+1)
gen_w1(nlin,k1,d)
#nlin=int(2*sqrt(2)*xmax*k2+1)
#gen_w2(nlin,k2,d)
gen_w2(nlin,k1,d)


fig_1=bqplot.Figure(marks=([make_map_1(E1,E2)]+list(w1_lin)+list(w2_lin)),
                                      layout=fig_layout_0, fig_margin=margin_0, axes=ax_x_y_1)

style = {'description_width': 'initial'}
sep_sldr=FloatSlider(min=0,max=10, step=0.1,description='Source separation',style=style,
                    continuous_update=False,value=d)
#k1_sldr=FloatSlider(min=0.1,max=5, step=0.1,description='\U0001d458\u2081',style=style,
k1_sldr=FloatSlider(min=0.1,max=5, step=0.1,description='Wave number',style=style,
                    continuous_update=False,value=k1)
#k2_sldr=FloatSlider(min=0.1,max=5, step=0.1,description='\U0001d458\u2082',style=style,
#                    continuous_update=False,value=k2)
ln_bttn=widgets.ToggleButton(value=True, description='Add wavefront lines')
hires=widgets.ToggleButton(value=False,
    description='High Resolution',
    disabled=False,
    button_style='danger',
    tooltip='Increases the number of points used for plotting. Selecting this will increase the time taken to render the figures.',
    icon='')

sep_sldr.observe(sep_updtr,names=['value'])
#k1_sldr.observe(E1_updtr,names=['value'])
#k2_sldr.observe(E2_updtr,names=['value'])
#k1_sldr.observe(E2_updtr,names=['value'])
k1_sldr.observe(E1_E2_updtr,names=['value'])
ln_bttn.observe(hide_updtr,names=['value'])
hires.observe(res_updtr,names=['value'])

sldr_box=widgets.VBox([widgets.HTML(value="<font size=5><strong>Spherical Wave Addition</strong></font>"),
                       widgets.HTML(value="<br>"),
                       widgets.HBox([hires,ln_bttn]),
                       widgets.HTML(value="<br>"),
#                       sep_sldr,k1_sldr,k2_sldr])
                       sep_sldr,k1_sldr])
display (widgets.HBox([fig_1,sldr_box]))

dt = timer() - start

print ("Plot generated in %f s" % dt)
```
